3) great! now we have the last big work to do : as we could see in the table, the dicilonerate could not use this type to do the regression.so we need to change it to an ordinal variable

first let's have a look at the column that we need to recode

```python
dis1.DisciplineRate
```

there are several problems we need to solve: 

first: values with "<"

```python
dis1.DisciplineRate[dis1.DisciplineRate.str.contains("<")]
```

then let the suppressed: N<10 becomes mission value

```python
import numpy as np
```

```python
dis1.replace(to_replace={'Suppressed: N<10':np.nan},inplace=True)
```

after solving those problems, we need to get the number in Disciplinerate: 

```python
dis1['number']=dis1.DisciplineRate.str.extract(pat='(\d+\.*\d*)')
dis1['number']
```

now, we create a flag, so we would remeber where we had a '<'

```python
dis1['flag_<']=dis1.DisciplineRate.str.contains("<")
dis1['flag_<']
```

then we comnibe the flag to the values

```python
list(zip(dis1.number,dis1['flag_<']))
```

them we need to substract 0.1 to every "number" where the "flag"is TURE

```python
[(float(number)-0.1) if flag==True else float(number) for number,flag in zip(dis1.number, dis1['flag_<'])]
```

then, we use the same to create a new column

```python
dis1['percent']=[(float(N)-0.1 if F else float(N)) for N,F in zip(dis1.number, dis1['flag_<'])]
```

```python
dis1['percent']
```

```python
dis1.percent.describe()
```

now after seeing the basic stats above, we creat intervals

```python
dis1['DisciplineInt']=pd.cut(dis1.percent,
                                  bins=[0,2,4,7,10,26],
                                  labels=['below_2','2_to_4','4_to_7','7_10','above_10'])
```

```python
dis1['DisciplineInt']
```

we could see the distribution of new variable

```python
dis1['DisciplineInt'].value_counts()
```

```python
dis1.iloc[:,-6:]
```

drop the ones we do not need

```python
to_drop=['number','flag_<','percent']
dis1.drop(columns=to_drop,inplace=True)
```

```python
dis1.head()
```

we are closed to the end! now we need to make discipline rate ordinal factor

```python
dis1['DisciplineInt'].value_counts()
```

```python
dis1['DisciplineInt']=pd.Categorical(dis1['DisciplineInt'],categories=["2_to_4","below_2","4_to_7","7_to_10","above_10"],ordered=True)
```
